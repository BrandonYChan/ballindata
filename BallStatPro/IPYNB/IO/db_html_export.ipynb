## Load Table

```python
db_path="sqlite:///../..//Ballbase.db" 
table_name= "Season_Stats_2023-24" 
engine=sqlalchemy.create_engine(db_path) 
df_from_db=pd.read_sql(table_name, con=engine)
df_from_db.sort_values('PTS', ascending=False)  
```

## Split Table

```python
# Total Stats Table 
totals = df_from_db.loc[:, 'Name':'FT_PCT'].sort_values('PTS', ascending=False) 

# Per Game Stats Table  
avgs = df_from_db.loc[:,['Name', 'Age', 'PPG', 'RPG', 'APG', 'SPG', 'BPG', 'ToPG',	'PFPG', 'MPG', 'FG_PCT', 'FG3_PCT',	'FT_PCT']].sort_values('PPG', ascending=False)  

totals 
avgs 

contents = [totals, avgs]
```

## Convert to HTML 

```python
# Create HTML table version 
for i in range(len(contents)):
    contents[i] = contents[i].to_html(table_id="table", classes = "table table-hover table-stripped table-bordered table-dark", index=False)

# Create page content 
for i in range(len(contents)):
    contents[i] = f"""
    {{%extends 'TableBase.html'%}}

    {{% block table %}}
    {contents[i]}
    {{% endblock %}}
"""
```

## Export to Page 

```python
def write_to_file(path, content): 
    file = open(path, "w", encoding="utf-8") 
    file.write(content) 
    file.close 


file_names=["Totals.html", "PerGame.html"] 

for i in range(len(contents)): 
    path = f"../templates/Tables/{file_names[i]}" 
    write_to_file(path, contents[i]) 
    
    
```
