## Imports

```python
import pandas as pd, numpy as np, sqlalchemy, sqlite3   
pd.set_option('display.max_columns', None)
```

## Implement Custom Function 

```python
def custom_function(df): 
    # Implement custom function 
    df = (df
          .drop(['-9999'], axis=1)  
          .fillna(0)
    )
    return df 
    
    
```

## Dataframes

```python
path = "../../static/CSV/allstars_2023_24.csv"

df = pd.read_csv(path) 
df = custom_function(df)

df 
```

## Export to Database

```python
db_path = "sqlite:///../../Ballbase.db"
# Naming convention: "Name_####-##" 
table_name = "AllstarGame_2023-24"
engine = sqlalchemy.create_engine(db_path, echo=False) 
df.to_sql(table_name, con=engine, if_exists="replace", index=False)
```
